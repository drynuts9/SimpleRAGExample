```python
import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader

# OpenAIのAPIキーを設定
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('rag')


# 1. データの読み込み (dataフォルダ内のファイルを自動認識)　基本情報技術者試験過去問
documents = SimpleDirectoryReader("./data").load_data()

# 2. インデックスの作成 (ベクトル化)
index = VectorStoreIndex.from_documents(documents)

# 3. クエリエンジンの作成と質問
query_engine = index.as_query_engine()
# response = query_engine.query("What's written in question1?")
# print(response)


# response = query_engine.query(
#     "From the document, find 'question 1'. Do not summarize."
# )
# print(response)

response = query_engine.query(
    # "何の書類が教えて。" A 基本情報技術者試験 科目 A の公開問題が教えています。
    "この文書は何ですか？簡潔に答えてください。"
)
print(response)

response = query_engine.query(
    "この文書は何？"
)
print(response)

response = query_engine.query(
    "ドキュメント内の『問１』『問２』『問３』の全文を、一字一句そのまま引用してください。要約や説明は禁止。"
)
print(response)

response = query_engine.query(
    "ドキュメント内にある『問１』『問２』『問３』それぞれの問題文と選択肢を、全文そのまま引用してください。見出しや説明ではなく、問題本文のみを出力してください。"
)
print(response)

response = query_engine.query(
    "ドキュメント内にある『問１』『問２』『問３』を回答してください。"
)
print(response)

response = query_engine.query(
    "ドキュメント内にある『問１』『問２』『問３』をkeyにして、内容をそのままvalueにしてJSONを作成してください"
)
print(response)



```
